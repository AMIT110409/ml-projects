```python
from tabulate import tabulate

arr=[]
purchased_amount=0;
for i in range(3):
  temp=[]
  temp.append(i+1)
  temp.append(input("Enter Item name"))
  qty=int(input("Enter qty"))
  amt=int(input("Enter amount"))
  temp.append(qty)
  temp.append(amt)
  purchased_amount+=qty*amt
  arr.append(temp)

df = tabulate(arr,  headers = ['SNo','Item-Name','Qty','Amount'])
dis=(purchased_amount*5)/100
gst=(purchased_amount*2)/100
total=purchased_amount+gst-dis
print("*****BILL******")
print(df)
print("Purchased amount", purchased_amount)
print("Discount", dis)
print("GST", gst)

print("-------------------")
print("Amount to be paid ",total)



```

```python
di={1:"A",2:"B",3:["JAVA","JSP","JAVASCRIPT"],4:{"s1":"py","s2":"var","s3":{"s4":"txt","s5":"doc"}}}
print(type(di))
keys=0;
key_value=0
ans=[]
for i in di:
  keys+=1
  key_value+=1
  if(type(di[i]) is tuple or type(di[i]) is list):
    for j in di[i]:
      ans.append(j)
  if(type(di[i]) is not dict and type(di[i]) is not tuple and type(di[i]) is not list):
    ans.append(di[i]) 
  if(type(di[i]) is dict):
    for j in di[i]:
      key_value+=1
      if(type(di[i][j]) is dict):
        for k in di[i][j]:
          key_value+=1
          ans.append(di[i][j][k])
      else:
        ans.append(di[i][j])
  
print("Total key-value pairs = ",key_value)
print("Total keys = ",keys)
print(ans)
```
