Importing the basic libraries

Importing the dataset

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv')
print(dataset.shape)
print(dataset.head())
```

```python
print(dataset.describe())
```

Data pre-processing (convert pandas Dataframe into a list )

```python
transactions = []
for i in range(0, 7500):
  transactions.append([str(dataset.values[i,j]) for  j  in range (0, 20)])
transactions = np.array(transactions) ## converting it into an numpy array

print(transactions.shape)## chechk the shape of the array

print(transactions[1:10]) ## having a look at the top customer's purchasing list 
```

TRaining APRIORI

```python
!pip install apyori
from apyori import apriori
```

```python
rules = apriori(transactions,min_support = 0.003, min_confidence = 0.2, min_lift=3,min_length=2)

results = list(rules)
results
```

Print out the no. of rules

```python
print(len(results))
```

have a glance at the rule

```python
print(results) ## glancing at the first rule
```
