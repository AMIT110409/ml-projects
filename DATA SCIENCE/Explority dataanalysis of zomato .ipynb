```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
## combining of dataset 

final_df=pd.merge(df,df_country,on='Country Code',how='left')   ## on means kis colum pe join kar rahe hai ..or how means jis o le rahe left yha right .. ko priority ..
```

```python
country_names=final_df.Country.value_counts().index 
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## pie chart  top 3 country
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%")
```

```python
#observation:Zomato maximum records or transcation are from India After that USA and then United Kingdoms 

rating=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
rating
```

```python
##conclusion
###
###1.when rating is between  4.5 to 4.9 --> Excellent
#2.When rating are between 4.0 to 3.4 --> very good 
3#. when rating is between 3.5 to 3.9 good 
#4. when rating is between 3.0 to 3.4 average
#4. when rating is between 2.5 to 2.9  average
#5. when rating is between 2.0 to 2.4  poor

```

```python
rating.head()
```

```python

import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count", data=rating)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count", data=rating,hue='Rating color',palette=['purple','red','orange','yellow','green','green'] )
```

2. maximum no  of rating are between 2.5 to 3.4
