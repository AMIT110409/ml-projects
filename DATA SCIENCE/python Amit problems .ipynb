```python
class madhyapradesh():
    def capital(self):
        print("bhopal is the capital of madhyapradesh.")
 
    def language(self):
        print("Malwa is the most widely spoken language of madhyapradesh.")
 
    def type(self):
        print("madhyapradesh is a developing state.")
 
class uttarpradesh():
    def capital(self):
        print("lucknow, lucknow is the capital of utterpradesh.")
 
    def language(self):
        print("Hindi is the primary language of USA.")
 
    def type(self):
        print("utterpradesh is a developed state.")
 
obj_mp = madhyapradesh()
obj_up = uttarpradesh()
for country in (obj_mp, obj_up):
    country.capital()
    country.language()
    country.type()
```

```python
class Bird:
  def intro(self):
    print("There are many types of birds.")
     
  def flight(self):
    print("Most of the birds can fly but some cannot.")
   
class Dove(Bird):
  def flight(self):
    print("Sparrows can fly.")
     
class ostrich(Bird):
  def flight(self):
    print("Ostriches cannot fly.")
     
obj_bird = Bird()
obj_spr = Dove()
obj_ost = ostrich()
 
obj_bird.intro()
obj_bird.flight()
 
obj_spr.intro()
obj_spr.flight()
 
obj_ost.intro()
obj_ost.flight()
```

```python
def fun(a):
    if a < 3:
 
        # throws ZeroDivisionError for a = 3
        b = a/(a-3)
 
    # throws NameError if a >= 4
    print("Value of b = ", b)
     
try:
    fun(4)
    fun(5)
 
# note that braces () are necessary here for
# multiple exceptions
except ZeroDivisionError:
    print("ZeroDivisionError Occurred and Handled")
except NameError:
    print("NameError Occurred and Handled")
```
